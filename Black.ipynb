##DATA CRAWLING

```python
import requests
import openpyxl
import pandas as pd
from bs4 import BeautifulSoup

df = pd.read_excel('Input.xlsx', nrows=114)

def extract_url_id(row):
  return row['URL'], row['URL_ID']

# Apply the function to each row and extract the URL and URL_ID
urls, url_ids = zip(*df.apply(extract_url_id, axis=1))
for url, url_id in zip(urls, url_ids):
        #Strip the url from empty new lines
    url_to_scrape = url.rstrip('\n')
    print('[FROM xlsx] Going to ' + url_to_scrape)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    page = requests.get(url_to_scrape,headers=headers)
    page_content = page.text
    soup = BeautifulSoup(page_content, 'lxml')
      
# Extract the article text 
    try:
        title = soup.find('header').text
        article_text = soup.find('div', class_="td-post-content").text
        # Save the article text to a text file 
        with open(f'{url_id}.txt', 'w') as f:  
          f.writelines([title, '\n\n', article_text])

    except:
      print("404 not found")
```

###SENTIMENT ANALYSIS

```python
# Mount Google Drive
import os
from google.colab import drive
drive.mount('/content/drive')
folder_path = '/content/drive/My Drive/20211030 Test Assignment/StopWords'
file_names = os.listdir(folder_path)
```

```python
import chardet

# empty list to store the stop words
stop_words = []

for file_name in file_names:
  with open(os.path.join(folder_path, file_name), 'rb') as f:
    raw_bytes = f.read()
  # Detect the encoding of the text
  encoding = chardet.detect(raw_bytes)['encoding']
  # Decode the text using the detected encoding
  text = raw_bytes.decode(encoding)
  stop_words += [line.strip() for line in text.split('\n')]

print(stop_words)
```

```python

# empty lists to store the positive and negative words
positive_words = []
negative_words = []

with open('positive-words.txt', 'rb') as f:
  raw_bytes = f.read()
  encoding = chardet.detect(raw_bytes)['encoding']
  text = raw_bytes.decode(encoding)
  positive_words += [line.strip() for line in text.split('\n')]

# Read the negative words
with open('negative-words.txt', 'rb') as f:
  raw_bytes = f.read()
  encoding = chardet.detect(raw_bytes)['encoding']
  text = raw_bytes.decode(encoding)
  negative_words += [line.strip() for line in text.split('\n')]

print(positive_words)
print(negative_words)
```

```python
import os
from textblob import TextBlob

drive.mount('/content/drive')
folder_path = '/content/drive/My Drive/txt'
file_names = os.listdir(folder_path)
text_file_names = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

# empty list to store the results of the text analysis
text_analysis_results = []

for file_name in text_file_names:
  with open(os.path.join(folder_path, file_name), 'r') as f:
    text = f.read()
  text = ' '.join([word for word in text.lower().split() if word not in stop_words])
  text_blob = TextBlob(text)
  # Extract the results of the text analysis
  positive_score = sum(word in positive_words for word in text.split())
  negative_score = sum(word in negative_words for word in text.split())
  polarity = text_blob.sentiment.polarity
  subjectivity = text_blob.sentiment.subjectivity

  text_analysis_results.append({'file_name': file_name, 'polarity': polarity, 'subjectivity': subjectivity, 'positive_score': positive_score, 'negative_score': negative_score})

print(text_analysis_results)

```

###Analysis of Readability

```python
drive.mount('/content/drive')
folder_path = '/content/drive/My Drive/txt'
file_names = os.listdir(folder_path)
text_file_names = [f for f in os.listdir(folder_path) if f.endswith('.txt')]
```

```python
import re
# Initialize an empty list to store the readability scores
readability_scores = []

# Iterate over the text files
for file_name in text_file_names:
  with open(os.path.join(folder_path, file_name), 'r') as f:
    text = f.read()
    text = ' '.join([word for word in text.lower().split() if word not in stop_words])
    words = text.split()
    sentences = text.split('.') 
    num_sentences = len(sentences)
    total_num_words = 0
    num_complex_words = 0
    total_num_syllables = 0
    total_num_pronouns = 0

  # Iterate over the sentences
    for sentence in sentences:
      words = sentence.split(' ')
      num_words = len(words)
      total_num_words += num_words
      for word in words:
      # If the word has more than two syllables, it is complex
        if len(word.split('-')) > 2:
          num_complex_words += 1
          word = re.sub(r'[^\w\s]', '', word)
        # Count the number of syllables in the word
          num_syllables = 0
          for vowel in 'aeiouAEIOU':
            num_syllables += word.count(vowel)
          # Check for exceptions where the word ends with "es" or "ed"
          if word.endswith('es'):
            num_syllables -= 1
          if word.endswith('ed'):
            num_syllables -= 1
          total_num_syllables += num_syllables
     # average number of syllables per word
        avg_syllables_per_word = total_num_syllables / total_num_words 
        if word.lower() in ['i', 'we', 'my', 'ours', 'us']:
          total_num_pronouns += 1

  avg_sentence_length = total_num_words / num_sentences
  percent_complex_words = num_complex_words / total_num_words * 100
  fog_index = 0.4 * (avg_sentence_length + percent_complex_words)

  readability_scores.append({'file_name': file_name, 'fog_index': fog_index, 'percent_complex_words' : percent_complex_words, 'avg_sentence_length': avg_sentence_length, 'Average syllables per word': avg_syllables_per_word, 'total_num_pronouns': total_num_pronouns,'complex_word_count':num_complex_words})
print(readability_scores)
```
